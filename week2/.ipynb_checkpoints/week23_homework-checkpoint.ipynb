# 4. uzdevums. Kodē kategoriskos mainīgos

## 4.1. Identificē kategoriskās kolonnas

```python
# Kuras kolonnas ir kategoriskas?
cat_cols = df.select_dtypes(include=['object', 'string', 'bool']).columns.tolist()

print("Kategoriskās kolonnas:", cat_cols)

# Parāda unikālās vērtības katrā kolonnā
for col in cat_cols:
    print(f"\n{col}:")
    print(df[col].unique())
```

## 4.2. Piemēro kodēšanu

Datu kopā ir vairākas kategoriskas kolonnas. Izmanto piemērotāko metodi:
• Label Encoding — kolonnām ar 2 vērtībām (piemēram, Weekend: True/False)
• One-Hot Encoding — kolonnām ar vairākām kategorijām (piemēram, Month, VisitorType)

```python
# Boolean kolonnas → 0/1
df['Weekend'] = df['Weekend'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)

# One-Hot Encoding
df = pd.get_dummies(df, columns=['Month', 'VisitorType'], drop_first=True)

# Ja One-Hot kolonnas ir bool, pārvērš tās uz int
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)

print(f"Kolonnu skaits pēc kodēšanas: {df.shape[1]}")
df.head()
```

```python
print(df.select_dtypes(include=['object', 'bool']).columns.tolist())
```

```python
df[['Month_Dec', 'VisitorType_Returning_Visitor']].dtypes
```

Kategoriskās kolonnas tika apstrādātas, pārveidojot tās skaitliskā formā:

* Kolonnas **`Weekend`** un **`Revenue`** tika pārvērstas no loģiskā tipa (`bool`) uz skaitliskām vērtībām (0 un 1), izmantojot `.astype(int)`.
* Kolonnām **`Month`** un **`VisitorType`** tika piemērots One-Hot Encoding ar `pd.get_dummies()`, izveidojot jaunas bināras kolonnas katrai kategorijas vērtībai (`drop_first=True`).

Pēc šīm darbībām sākotnējās kategoriskās kolonnas tika aizstātas ar skaitliskām, un datu kopā vairs nepalika `object` tipa kolonnas.


## 4.3. Pārbaude

```python
# Pārliecinies, ka nav palikušas 'object' vai 'bool' kolonnas
remaining_objects = df.select_dtypes(include=['object', 'bool']).columns.tolist()

print("Atlikušās kategoriskās kolonnas:", remaining_objects)

if len(remaining_objects) == 0:
    print("✓ Visas kategoriskās kolonnas ir veiksmīgi kodētas.")
else:
    print("✗ Ir neapstrādātas kategoriskās kolonnas! Nepieciešama papildu kodēšana.")

assert len(remaining_objects) == 0, "Vēl ir neapstrādātas kategoriskas kolonnas!"
```

# 5. uzdevums. Izveido jaunu iezīmi (Feature Engineering)

## 5.1. Jauna feature izveide

```python
# Piemērs 1: Kopējais laiks vietnē
df['TotalDuration'] = (df['Administrative_Duration'] +
df['Informational_Duration'] +
df['ProductRelated_Duration'])
```

```python
# Piemērs 2: Bounce-to-Exit attiecība
df['BounceExitRatio'] = df['BounceRates'] / df['ExitRates']
df.loc[df['ExitRates'] == 0, 'BounceExitRatio'] = np.nan

df['BounceExitRatio'] = df['BounceRates'] / (df['ExitRates'] + 0.001) # Dotais
```

```python
# Piemērs 3: Produktu pārlūkošanas intensitāte
df['ProductBrowsingIntensity'] = df['ProductRelated'] / (df['ProductRelated_Duration'] + 0.001)
```

## 5.2. Paskaidro jauno feature

Jaunais mainīgais **TotalDuration** tika izveidots, summējot trīs dažādu aktivitāšu laikus:
- Administrative Duration
- Informational Duration
- Product Related Duration

Šī iezīme atspoguļo kopējo lietotāja pavadīto laiku vietnē. 
Kopējais pavadītais laiks vietnē var norādīt uz lietotāja iesaisti (engagement). Lietotāji, kuri pavada vairāk laika, bieži:
aktīvāk pārlūko produktus, salīdzina piedāvājumus, un ir lielāka iespēja, ka viņi veiks pirkumu.

Noderība ML modelim. Šī feature apvieno vairākas atsevišķas kolonnas vienā informatīvā rādītājā, kas: 
- samazina datu dimensiju, 
- var palīdzēt modelim labāk uztvert lietotāja uzvedības kopējo intensitāti, 
- potenciāli uzlabo prognozēšanas precizitāti attiecībā uz Revenue.


Jaunais mainīgais **BounceExitRatioauns** raksturo attiecību starp lietotāju tūlītēju aiziešanu no lapas (bounce) un lapas pamešanu kopumā (exit).
Šis rādītājs palīdz labāk izprast lietotāju uzvedību:
- Augsta vērtība norāda, ka liela daļa lietotāju pamet lapu uzreiz pēc ielādes (zema iesaiste).
- Zema vērtība liecina, ka lietotāji mijiedarbojas ar lapu un aiziet vēlāk (augstāka iesaiste).
Salīdzinot ar sākotnējiem mainīgajiem (BounceRates un ExitRates), šī attiecība sniedz papildu informāciju par lietotāju uzvedības struktūru un var uzlabot prognozēšanas modeļu precizitāti.


Jaunais mainīgais **ProductBrowsingIntensity** mēra lietotāja pārlūkošanas aktivitāti, 
aprēķinot apmeklēto produktu lapu skaitu attiecībā pret tam veltīto laiku. 
Augsta vērtība norāda uz intensīvu, ātru pārlūkošanu, kamēr zema vērtība liecina par lēnāku un padziļinātāku produktu izpēti.


# 5.3. Vizualizē jauno feature

```python
# Parādi, kā jaunā iezīme atšķiras starp pircējiem un nepircējiem
plt.figure(figsize=(8, 4))
df.groupby('Revenue')['TotalDuration'].mean().plot(kind='bar',
color=['steelblue', 'coral'])
plt.title('Vidējais kopējais laiks vietnē: pircēji vs. nepircēji')
plt.ylabel('TotalDuration (normalizēta)')
plt.xticks([0, 1], ['Nav pirkuma', 'Ir pirkums'], rotation=0)
plt.tight_layout()
plt.show()
```

Grafikā tiek salīdzināts vidējais kopējais pavadītais laiks vietnē **TotalDuration** starp lietotājiem, kuri veica pirkumu, un tiem, kuri pirkumu neveica.
No rezultātiem var novērot, ka lietotāji, kuri veic pirkumu, parasti pavada vairāk laika vietnē. Tas liecina, ka lielāka iesaiste un ilgāka mijiedarbība ar vietni var būt saistīta ar lielāku pirkuma varbūtību.
Šāda vizualizācija palīdz novērtēt jaunās iezīmes informatīvo vērtību un tās potenciālo nozīmi prognozēšanas modelī.

```python
# Histrogramma BounceExitRatio
import matplotlib.pyplot as plt
plt.hist(df['BounceExitRatio'], bins=50)
plt.title('BounceExitRatio sadalījums')
plt.xlabel('BounceExitRatio')
plt.ylabel('Biežums')
plt.show()
```

Histogramma rāda, ka lielākā daļa **BounceExitRatio vērtību** ir tuvu nullei, kas nozīmē,

Divi pīķi pie 0 un 1 norāda uz atšķirīgām lietotāju grupām — vieni ātri aiziet, citi ilgāk mijiedarbojas. 
Augstākas vērtības ir retas un atspoguļo ekstrēmus gadījumus ar ļoti zemu ExitRates

- ProductRelated: produktu lapu apmeklējumu skaits
- ProductRelated_Duration: laiks produktu lapās
- BounceRates: atlēcienu procents sesijā
- ExitRates: pamešanas procents sesijā
- PageValues: Google Analytics lapas vērtība sesijā
- SpecialDay: tuvums īpašai dienai (piem., svētki)
Kategoriskie un binārie rādītāji
- Month: sesijas mēnesis
- OperatingSystems: lietotāja OS kods
- Browser: pārlūka kods
- Region: reģiona kods
- TrafficType: satiksmes avots (no kāda avota lietotājs ieradās mājaslapā sesijas laikā)
    - tiešais apmeklējums (lietotājs ieradās tieši lapā, piemēram, ierakstot URL)
    - meklētājprogrammas (Google, Bing)
    - reklāmas kampaņas, u.c.
- VisitorType: jauns vai atkārtots apmeklētājs
- Weekend: sesija brīvdienā (TRUE/FALSE)
- Revenue: mērķis — vai sesija beidzās ar pirkumu (TRUE/FALSE)

Datu bāzē nav iztrūkstošo vērtību.

Datu kopā tika konstatētas kategoriskās (`object`) un loģiskās (`bool`) kolonnas, kuras nevar tieši izmantot mašīnmācīšanās modeļos.
Lai atrisinātu šīs problēmas, kategoriskie mainīgie tika pārveidoti, izmantojot One-Hot Encoding, bet boolean kolonnas tika konvertētas uz skaitliskām (0 un 1). 
Pēc apstrādes visi mainīgie ir skaitliskā formātā.
Datu kopas dimensijas mainījās — kolonnu skaits palielinājās no sākotnējā kolonnu skaita 18 uz 27 kolonnām, 
jo One-Hot Encoding izveidoja papildu kolonnas no kategoriskajiem mainīgajiem.
