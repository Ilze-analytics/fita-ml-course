# 4. uzdevums - Definē ML problēmu

## **A. Problēmas apraksts**

**Problēma:** Prognozēt, vai pasažieris izdzīvos Titānika katastrofā, balstoties uz dzimumu, vecumu, biļetes klasi un citiem datiem.  

**Svarīgums:** Palīdz uzlabot drošības un evakuācijas procedūras, identificējot riska grupas.  

**Gala lietotājs:** Kuģa drošības komanda un analītiķi, kas veido evakuācijas plānus.  

**Mērķis:** Saprast, kuri faktori ietekmē izdzīvošanu un aprēķināt izdzīvošanas varbūtību.


## **B. ML uzdevuma tips**

**Uzdvuma veids:** Supervised learning  
- Tas ir supervised, jo mums ir zināmas **mērķa vērtības** (`survived`: 1 = izdzīvoja, 0 = neizdzīvoja). Modeļi tiek trenēti, lai prognozētu šo mērķa vērtību, izmantojot citus datus (features).

**Uzdevuma tips:** Klasifikācija  
- Tas ir klasifikācijas uzdevums, jo mērķis ir prognozēt **kategorisku iznākumu** — vai pasažieris izdzīvos (1) vai neizdzīvos (0).


## **C. Target variable**

Šajā sadaļā definēsim **target mainīgo**, tas ir, vērtību, ko mēģinām prognozēt ar mūsu ML modeli.  
Šajā uzdevumā target mainīgais ir `survived`, kas norāda, vai pasažieris izdzīvoja (1) vai neizdzīvoja (0).

```python
# Target mainīgais
print(df['survived'].value_counts())  # cik pasažieru izdzīvoja/neizdzīvoja
print(f'\nIzdzīvoja: {df["survived"].mean():.1%}')  # procentuāli
```

**Paskaidrojums:**  

**Target mainīgais (`survived`)** ir mūsu mērķis, ko mēģinām prognozēt datu analīzē vai mašīnmācīšanās uzdevumā.

**Kolonna `survived`:**  
- Kolonna `survived` ir target mainīgais, ko mēģinām prognozēt.  
- Vērtības:  
  - `0` – pasažieris neizdzīvoja  
  - `1` – pasažieris izdzīvoja

**Lietojums Python kodā:**  
- `df['survived']` - kolonna “survived” datu rāmī `df`.  
- `df['survived'].value_counts()` - parāda, cik cilvēki izdzīvoja (1) un cik neizdzīvoja (0).  
- `df["survived"].mean()` - aprēķina procentuālo daļu izdzīvojušo:  
  - Ja vērtības ir 0 un 1, vidējā vērtība = izdzīvojušo īpatsvars.  
  - `:.1%` → formatē skaitli procentos ar vienu zīmi aiz komata.

## **D. Features saraksts**

## Features (iezīmes) datu modelēšanai

### Skaitliskās (numerical)
- age - cilvēka vecums
- fare - biļetes cena
- sibsp - brāļu/māsu un partneru skaits
- parch - vecāku/bērnu skaits

### Kategoriskās (categorical)
- sex - dzimums
- pclass - kuģa klase
- embarked - iekāpšanas osta
- alone - vai cilvēks ceļoja viens



```python
# Features pārskats
print('Skaitliskās features:')
print(df.select_dtypes(include='number').columns.tolist())
# select_dtypes(include='number') - atlasīs visas numeriskās kolonnas
print('\nKategoriskās features:')
print(df.select_dtypes(include='category').columns.tolist())
# select_dtypes(include='category') - atlasīs visas kategoriskās kolonnas
# columns.tolist() - izdrukā kolonnas kā sarakstu
```

**Paskaidrojums:**  

- **Skaitliskās features** – mainīgie ar skaitliskām vērtībām, kur var veikt aprēķinus un statistiku.  
- **Kategoriskās features** – mainīgie, kas satur kategorijas vai grupas, piemēram, dzimums vai biļetes klase.  
- ML modelī šīs features tiek izmantotas, lai prognozētu target mainīgo (`survived`).
